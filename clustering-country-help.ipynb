## 3. Outlier treatment

check outlier

```python
sns.set({'figure.figsize':(10,6)})
sns.boxplot(data=df, x='Pendapatan')
```

```python
sns.set({'figure.figsize':(10,6)})
sns.boxplot(data=df, x='GDPperkapita')
```

```python
def remove_outlier(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3-q1
    df_final = df[~(df>(q1-(1.5*iqr))) | (df<(q3+(1.5*iqr)))]
#     df_final = df[(df>(q1-(1.5*iqr))) & (df<(q3+(1.5*iqr)))]
    return df_final
```

```python
df2 = remove_outlier(df[['Pendapatan', 'GDPperkapita']])
df2.dropna(axis=0, inplace=True)
df2
```

```python
sns.set({'figure.figsize':(10,6)})
sns.boxplot(data=df2, x='Pendapatan')
```

```python
sns.set({'figure.figsize':(10,6)})
sns.boxplot(data=df2, x='GDPperkapita')
```

## 4. Scaling data

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df_std = ss.fit_transform(df2)
```

```python
from sklearn.cluster import KMeans

kmeans1 = KMeans(n_clusters=2,random_state=42)
kmeans1.fit(df_std)
labels1 = kmeans1.labels_
labels1
```

```python
new_df = pd.DataFrame(data=df_std, columns=['Pendapatan','GDPperkapita'])
new_df['labels1'] = labels1
new_df
```

```python
plt.figure(figsize=(12,8))

plt.scatter(new_df['Pendapatan'][new_df.labels1==0],
            new_df['GDPperkapita'][new_df.labels1==0],
            c='red',s=100,edgecolors='yellow',label='Negara butuh bantuan')
plt.scatter(new_df['Pendapatan'][new_df.labels1==1],
            new_df['GDPperkapita'][new_df.labels1==1],
            c='green',s=100,edgecolors='yellow',label='Negara tidak butuh bantuan')

plt.scatter(kmeans1.cluster_centers_[:,0], kmeans1.cluster_centers_[:,0], c='k', s=300)

plt.legend()
plt.show()
```

## Using elbow method to find optimal number of clusters

```python
wcss = []
for i in range(1,11):
    kMeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kMeans.fit(df_std)
    wcss.append(kMeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Num. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans2 = KMeans(n_clusters=3,random_state=42,init='k-means++')
kmeans2.fit(df_std)
labels2 = kmeans2.labels_
labels2
```

```python
new_df['labels2'] = labels2
new_df
```

```python
plt.figure(figsize=(12,8))

plt.scatter(new_df['Pendapatan'][new_df.labels2==0],
            new_df['GDPperkapita'][new_df.labels2==0],
            c='red',s=100,edgecolors='yellow',label='0')
plt.scatter(new_df['Pendapatan'][new_df.labels2==1],
            new_df['GDPperkapita'][new_df.labels2==1],
            c='green',s=100,edgecolors='yellow',label='1')
plt.scatter(new_df['Pendapatan'][new_df.labels2==2],
            new_df['GDPperkapita'][new_df.labels2==2],
            c='yellow',s=100,edgecolors='yellow',label='2')

plt.scatter(kmeans2.cluster_centers_[:,0], kmeans2.cluster_centers_[:,1], c='k', s=300)

plt.legend()
plt.show()
```

silhoutte score

```python
from sklearn.metrics import silhouette_score

print('Score n_clusters = 2:',silhouette_score(df_std, labels= labels1))
print('Score n_clusters = 3:',silhouette_score(df_std, labels= labels2))
```
